```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a Pandas DataFrame with the provided data
data = {'category': ['Beauty', 'Beauty', 'Clothing', 'Clothing', 'Electronics', 'Electronics'],
        'gender': ['Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'total_trans': [330, 281, 347, 351, 343, 335]}
df = pd.DataFrame(data)

# Group the data by category and gender
grouped_data = df.groupby(['category', 'gender'])['total_trans'].sum().reset_index()

# Create the grouped bar chart
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figure size as needed

width = 0.35  # Width of each bar

# Get unique categories and genders
categories = grouped_data['category'].unique()
genders = grouped_data['gender'].unique()

# Create bars for each gender within each category
for i, gender in enumerate(genders):
    x_pos = [j + i * width for j in range(len(categories))]
    values = grouped_data[grouped_data['gender'] == gender]['total_trans'].values
    ax.bar(x_pos, values, width, label=gender)

# Set x-axis labels and ticks
ax.set_xticks([j + width / 2 for j in range(len(categories))])
ax.set_xticklabels(categories)

# Set labels and title
ax.set_xlabel("Category")
ax.set_ylabel("Total Transactions")
ax.set_title("Grouped Bar Chart of Total Transactions by Category and Gender")

# Add legend
ax.legend()

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a Pandas DataFrame with the provided data
data = {'category': ['Beauty', 'Beauty', 'Clothing', 'Clothing', 'Electronics', 'Electronics'],
        'gender': ['Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'total_trans': [330, 281, 347, 351, 343, 335]}
df = pd.DataFrame(data)

# Group the data by category and gender
grouped_data = df.groupby(['category', 'gender'])['total_trans'].sum().reset_index()

# Create the grouped bar chart
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figure size as needed

width = 0.35  # Width of each bar

# Get unique categories and genders
categories = grouped_data['category'].unique()
genders = grouped_data['gender'].unique()

# Create bars for each gender within each category
for i, gender in enumerate(genders):
    x_pos = [j + i * width for j in range(len(categories))]
    values = grouped_data[grouped_data['gender'] == gender]['total_trans'].values
    ax.bar(x_pos, values, width, label=gender)

# Set x-axis labels and ticks
ax.set_xticks([j + width / 2 for j in range(len(categories))])
ax.set_xticklabels(categories)

# Set labels and title
ax.set_xlabel("Category")
ax.set_ylabel("Total Transactions")
ax.set_title("Grouped Bar Chart of Total Transactions by Category and Gender")

# Add legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Add count labels on top of bars
for i, gender in enumerate(genders):
    x_pos = [j + i * width for j in range(len(categories))]
    values = grouped_data[grouped_data['gender'] == gender]['total_trans'].values
    for x, y in zip(x_pos, values):
        ax.text(x, y + 2, str(y), ha='center', va='bottom')  # Adjust position as needed

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a Pandas DataFrame with the provided data
data = {'customer_id': [3, 1, 5, 2, 4],
        'total_sales': [38440, 30750, 30405, 25295, 23580]}
df = pd.DataFrame(data)

# Create the bar chart
plt.figure(figsize=(8, 6))  # Adjust figure size as needed

# Get the index of the top customer
top_customer_index = df['total_sales'].idxmax()

# Create a list of colors, with red for the top customer
colors = ['orange'] * len(df)  # Default color for all bars
colors[top_customer_index] = 'red'  # Set red for the top customer

# Create the bar chart with custom colors
plt.bar(df['customer_id'], df['total_sales'], color=colors)

# Set labels and title
plt.xlabel("Customer ID")
plt.ylabel("Total Sales")
plt.title("Total Sales by Customer")

# Add count labels on top of bars
for i, v in enumerate(df['total_sales']):
    plt.text(df['customer_id'][i], v + 500, str(v), ha='center', va='bottom')

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for unique customer count by category
data = {'category': ['Beauty', 'Clothing', 'Electronics'],
        'cnt_unique_cs': [141, 149, 144]}
df = pd.DataFrame(data)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Bar chart
axes[0].bar(df['category'], df['cnt_unique_cs'], color=['skyblue', 'lightgreen', 'gold'])
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Unique Customer Count")
axes[0].set_title("Unique Customers by Category (Bar Chart)")

# Add count labels on top of bars
for i, v in enumerate(df['cnt_unique_cs']):
    axes[0].text(df['category'][i], v + 0.5, str(v), ha='center', va='bottom')

# Pie chart
axes[1].pie(df['cnt_unique_cs'], labels=df['category'], autopct='%1.1f%%',
            startangle=90, colors=['skyblue', 'lightgreen', 'gold'])
axes[1].set_title("Unique Customers by Category (Pie Chart)")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for total profit by category
data = {'category': ['Clothing', 'Electronics', 'Beauty'],
        'total_profit': [245945.2499999999, 244767.54999999996, 228589.3999999999]}
df = pd.DataFrame(data)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Bar chart with rounded profit values
axes[0].bar(df['category'], df['total_profit'], color=['skyblue', 'lightcoral', 'gold'])
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Total Profit")
axes[0].set_title("Total Profit by Category (Bar Chart)")

# Add rounded count labels on top of bars
for i, v in enumerate(df['total_profit']):
    axes[0].text(df['category'][i], v + 2000, f"{v:.0f}", ha='center', va='bottom')

# Pie chart with rounded profit values in labels
axes[1].pie(df['total_profit'], labels=df['category'], autopct='%1.1f%%',
            startangle=90, colors=['skyblue', 'lightcoral', 'gold'])
axes[1].set_title("Total Profit by Category (Pie Chart)")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for category revenue and revenue percentage
data = {'category': ['Electronics', 'Clothing', 'Beauty'],
        'category_revenue': [311445, 309995, 286790],
        'revenue_percentage': [34.291423978507645, 34.13177278883102, 31.576803232661334]}
df = pd.DataFrame(data)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Bar chart for category revenue
axes[0].bar(df['category'], df['category_revenue'], color=['skyblue', 'lightcoral', 'gold'])
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Category Revenue")
axes[0].set_title("Category Revenue")

# Add count labels on top of bars
for i, v in enumerate(df['category_revenue']):
    axes[0].text(df['category'][i], v + 2000, f"{v:.0f}", ha='center', va='bottom')

# Pie chart for revenue percentage
axes[1].pie(df['revenue_percentage'], labels=df['category'], autopct='%1.1f%%',
            startangle=90, colors=['skyblue', 'lightcoral', 'gold'])
axes[1].set_title("Revenue Percentage")

# Adjust layout
plt.tight_layout()
plt.show()
```
